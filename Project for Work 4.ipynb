```python
from datetime import datetime

date_strings = ['6-3-1980', '12-8-1983', '21-3-1975', '5-2-1975', '5-4-1985', '29-11-1980', '10-1-1979',
                '27-11-1983', '11-6-1984', '6-11-1977', '30-11-1983', '16-8-1990', '11-7-1981', '2-5-1981',
                '18-5-1983', '18-9-1977', '29-9-1977', '12-10-1990', '15-9-1975', '16-8-1977', '14-11-1975',
                '18-1-1987', '17-7-1988', '11-11-1980', '21-10-1990', '27-7-1989', '14-11-1975', '19-10-1977',
                '22-7-1986', '25-7-1983']

dates = [datetime.strptime(date_str, '%d-%m-%Y') for date_str in date_strings]

current_date = datetime(2024, 6, 24)
ages = [(current_date - date).days // 365 for date in dates]

plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, edgecolor='black', alpha=0.7, color = 'Green')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.grid(False)
plt.show()
```

```python
df = pd.DataFrame(data)
company_counts = df['Company'].value_counts()
explode = (0 , 0, 0, 0, 0.1)
plt.figure(figsize=(8, 6))
plt.pie(company_counts, labels=company_counts.index, explode=explode, autopct='%1.1f%%', startangle=140,
        shadow = True, colors=sns.color_palette('viridis', len(company_counts)))
plt.title('Company')
plt.axis('equal')
plt.show()
```

```python
df = pd.DataFrame(data)

df_3d_real_design = df[df['Company'] == '3D Real Design']
explode = (0.1 , 0, 0, 0, 0)
position_counts = df_3d_real_design['Position'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140,
         explode = explode, shadow = True, colors=sns.color_palette('viridis', len(position_counts)))
plt.title('Position of the company "3D Real Design"')
plt.axis('equal')
plt.show()
```
